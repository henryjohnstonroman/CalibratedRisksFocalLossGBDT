```python
# Load malignancy data (malig) 
malig1 = pd.read_csv('Data/SRTR/malig.csv')
malig1['TRR_FOL_ID'] = malig1['TRR_FOL_ID'].astype('object')
malig1.shape
```

```python
# Match patients in follow-up file to patients in malignancy file
malig2 = malig1.loc[malig1['TRR_FOL_ID'].isin(txf_lu3['TRR_FOL_ID'])]
malig2.shape
```

```python
# Remove patients without a PTLD pathology date
malig3 = malig2.loc[~malig2['MAL_PTLD_DGN_DT'].isnull()]
malig3.shape
```

```python
# Merge follow-up data (txf_lu3) with malignancy file (malig3) so we can compute time-to-event (time-to-PTLD)
# Let's call it "ptld" (PTLD(+) patients)

ptld = pd.merge(txf_lu3, malig3, on="TRR_FOL_ID")
ptld.shape
```

```python
# Compute time-to-event for PTLD(+) patients
# pd.set_option('display.max_rows', None)
time_to_PTLD = pd.to_datetime(ptld['MAL_PTLD_DGN_DT'])-pd.to_datetime(ptld['REC_TX_DT'])
ptld['TIME_TO_EVENT'] = time_to_PTLD

# PTLD(+) patients in one year
ptld1Y = ptld[ptld['TIME_TO_EVENT']<pd.Timedelta(365*1, unit='d')]
ptld1Y = ptld1Y.drop_duplicates('PERS_ID') # Check for duplicates
print(ptld1Y.shape)

# PTLD(+) patients in two years
ptld2Y = ptld[ptld['TIME_TO_EVENT']<pd.Timedelta(365*2, unit='d')]
ptld2Y = ptld2Y.drop_duplicates('PERS_ID') # Check for duplicates
print(ptld2Y.shape)

# PTLD(+) patients in three years
ptld3Y = ptld[ptld['TIME_TO_EVENT']<pd.Timedelta(365*3, unit='d')]
ptld3Y = ptld3Y.drop_duplicates('PERS_ID') # Check for duplicates
print(ptld3Y.shape)

# PTLD(+) patients in four years
ptld4Y = ptld[ptld['TIME_TO_EVENT']<pd.Timedelta(365*4, unit='d')]
ptld4Y = ptld4Y.drop_duplicates('PERS_ID') # Check for duplicates
print(ptld4Y.shape)

# PTLD(+) patients in five years
ptld5Y = ptld[ptld['TIME_TO_EVENT']<pd.Timedelta(365*5, unit='d')]
ptld5Y = ptld5Y.drop_duplicates('PERS_ID') # Check for duplicates
print(ptld5Y.shape)

# PTLD(+) patients in eight years
ptld8Y = ptld[ptld['TIME_TO_EVENT']<pd.Timedelta(365*8, unit='d')]
ptld8Y = ptld8Y.drop_duplicates('PERS_ID') # Check for duplicates
print(ptld8Y.shape)

# PTLD(+) patients in 10 years
ptld10Y = ptld[ptld['TIME_TO_EVENT']<pd.Timedelta(365*10, unit='d')]
ptld10Y = ptld10Y.drop_duplicates('PERS_ID') # Check for duplicates
print(ptld10Y.shape)

# PTLD(+) patients in 12 years
ptld12Y = ptld[ptld['TIME_TO_EVENT']<pd.Timedelta(365*12, unit='d')]
ptld12Y = ptld12Y.drop_duplicates('PERS_ID') # Check for duplicates
print(ptld12Y.shape)

# PTLD(+) patients in 15 years
ptld15Y = ptld[ptld['TIME_TO_EVENT']<pd.Timedelta(365*15, unit='d')]
ptld15Y = ptld15Y.drop_duplicates('PERS_ID') # Check for duplicates
print(ptld15Y.shape)


# Check for duplicates in PTLD patients, PTLD(+) all
ptld = ptld.sort_values(by=['PERS_ID', 'TIME_TO_EVENT'])
ptld = ptld.drop_duplicates('PERS_ID') # Check for duplicates
print(ptld.shape)
```

```python
# Extract ids of PTLD patients with and without diagnosis date in the malignancy file
allPTLDID = txf_lu3[txf_lu3['TFL_MALIG_LYMPH']=='Y']['PERS_ID'].drop_duplicates()
ptld_nodateID = allPTLDID[~allPTLDID.isin(ptld['PERS_ID'])]
ptld_nodate = txf_lu3[txf_lu3['PERS_ID'].isin(ptld_nodateID)]
ptld_nodate = ptld_nodate[ptld_nodate['TFL_MALIG_LYMPH']=='Y']
```

```python
# Compute time-to-event for PTLD(+) patients without diagnosis date. 
# Use first follow-up date at which malignancy was identified.
time_to_PTLD2 = pd.to_datetime(ptld_nodate['TFL_PX_STAT_DT'])-pd.to_datetime(ptld_nodate['REC_TX_DT'])
ptld_nodate['TIME_TO_EVENT'] = time_to_PTLD2

# PTLD(+) patients in one year
ptld1Y2 = ptld_nodate[ptld_nodate['TIME_TO_EVENT']<pd.Timedelta(365*1, unit='d')]
ptld1Y2 = ptld1Y2.drop_duplicates('PERS_ID') # Check for duplicates
print(ptld1Y2.shape)

# PTLD(+) patients in two years
ptld2Y2 = ptld_nodate[ptld_nodate['TIME_TO_EVENT']<pd.Timedelta(365*2, unit='d')]
ptld2Y2 = ptld2Y2.drop_duplicates('PERS_ID') # Check for duplicates
print(ptld2Y2.shape)

# PTLD(+) patients in three years
ptld3Y2 = ptld_nodate[ptld_nodate['TIME_TO_EVENT']<pd.Timedelta(365*3, unit='d')]
ptld3Y2 = ptld3Y2.drop_duplicates('PERS_ID') # Check for duplicates
print(ptld3Y2.shape)

# PTLD(+) patients in four years
ptld4Y2 = ptld_nodate[ptld_nodate['TIME_TO_EVENT']<pd.Timedelta(365*4, unit='d')]
ptld4Y2 = ptld4Y2.drop_duplicates('PERS_ID') # Check for duplicates
print(ptld4Y2.shape)

# PTLD(+) patients in five years
ptld5Y2 = ptld_nodate[ptld_nodate['TIME_TO_EVENT']<pd.Timedelta(365*5, unit='d')]
ptld5Y2 = ptld5Y2.drop_duplicates('PERS_ID') # Check for duplicates
print(ptld5Y2.shape)

# PTLD(+) patients in eight years
ptld8Y2 = ptld_nodate[ptld_nodate['TIME_TO_EVENT']<pd.Timedelta(365*8, unit='d')]
ptld8Y2 = ptld8Y2.drop_duplicates('PERS_ID') # Check for duplicates
print(ptld8Y2.shape)

# PTLD(+) patients in 10 years
ptld10Y2 = ptld_nodate[ptld_nodate['TIME_TO_EVENT']<pd.Timedelta(365*10, unit='d')]
ptld10Y2 = ptld10Y2.drop_duplicates('PERS_ID') # Check for duplicates
print(ptld10Y2.shape)

# PTLD(+) patients in 12 years
ptld12Y2 = ptld_nodate[ptld_nodate['TIME_TO_EVENT']<pd.Timedelta(365*12, unit='d')]
ptld12Y2 = ptld12Y2.drop_duplicates('PERS_ID') # Check for duplicates
print(ptld12Y2.shape)

# PTLD(+) patients in 15 years
ptld15Y2 = ptld_nodate[ptld_nodate['TIME_TO_EVENT']<pd.Timedelta(365*15, unit='d')]
ptld15Y2 = ptld15Y2.drop_duplicates('PERS_ID') # Check for duplicates
print(ptld15Y2.shape)


# Check for duplicates in PTLD patients, PTLD(+) all
ptld_nodate = ptld_nodate.sort_values(by=['PERS_ID', 'TIME_TO_EVENT'])
ptld_nodate = ptld_nodate.drop_duplicates('PERS_ID') # Check for duplicates
print(ptld_nodate.shape)
```

```python
# Merge immunosuppression data with tx_lu data
immuno = pd.read_csv('Data/SRTR/immuno.csv')
immuno = immuno.replace(0, np.nan)
immuno['REC_DRUG_CD'] = immuno['REC_DRUG_CD'].replace([-2,3,4,44,46,48],333)
immuno['REC_DRUG_CD'] = immuno['REC_DRUG_CD'].replace([5,53,54,59],555)
immuno['REC_DRUG_CD'] = immuno['REC_DRUG_CD'].replace([9,47,55,57],999)
immuno['REC_DRUG_CD'] = immuno['REC_DRUG_CD'].replace([6,58],666)
immuno['REC_DRUG_CD'] = immuno['REC_DRUG_CD'].replace([1,2,49],4949)
IND = immuno['REC_DRUG_CD']*immuno['REC_DRUG_INDUCTION']
ANTI_REJ = immuno['REC_DRUG_CD']*immuno['REC_DRUG_ANTI_REJ']
MAINT = immuno['REC_DRUG_CD']*immuno['REC_DRUG_MAINT']
immuno_set = pd.concat([immuno['TRR_ID'], pd.get_dummies(IND, prefix='INDUCTION')],axis=1)
immuno_set = pd.concat([immuno_set, pd.get_dummies(ANTI_REJ, prefix='ANTI_REJECTION')],axis=1)
immuno_set = pd.concat([immuno_set, pd.get_dummies(MAINT, prefix='MAINTENANCE_DISCHARGE')],axis=1)
immuno_set = immuno_set.groupby('TRR_ID').sum()
immuno_set = immuno_set.replace([2,3,4],1)
```
